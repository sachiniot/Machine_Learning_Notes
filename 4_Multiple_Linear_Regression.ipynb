```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = {
    "Size":[1200,1500,1800,2000],
    "Bedrooms":[2,3,3,4],
    "Age":[5,3,2,1],
    "Price":[200000,250000,300000,350000]
}

df = pd.DataFrame(data)

X = df[['Size','Bedrooms','Age']]
y = df['Price']

model = LinearRegression()

model.fit(X,y)

prediction = model.predict([[1600,3,2]])

print(prediction)
```
