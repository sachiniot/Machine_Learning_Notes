```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Step 1: Create dataset
data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male'],
    'City': ['Delhi', 'Mumbai', 'Chennai', 'Delhi', 'Mumbai'],
    'Salary': [50000, 60000, 55000, 52000, 58000]
}

df = pd.DataFrame(data)

# Step 2: Convert categorical variables to dummy variables
df_encoded = pd.get_dummies(df, drop_first=True) # dropping first column to ignore dummy varaible trap

# Step 3: Define X (features) and y (target)
X = df_encoded.drop('Salary', axis=1)
y = df_encoded['Salary']

# Step 4: Train regression model
model = LinearRegression()
model.fit(X, y)

# Step 5: See results
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```
