```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Data (exponential)
X = np.array([1, 2, 3, 4, 5]).reshape(-1,1)
y = np.array([2, 4, 8, 16, 32])

# BEFORE (original)
model = LinearRegression()
model.fit(X, y)
y_pred_before = model.predict(X)

print("R2 Before:", r2_score(y, y_pred_before))

plt.figure()
plt.scatter(X, y)
plt.plot(X, y_pred_before)
plt.title("Before Transformation")
plt.show()


# ✅ AFTER (log transform on y)
# here I am transforming Y not X because in this exponential occurs in Y not in X.
# Apply transformation to that which contain exponential not necessarily always on X.
y_log = np.log(y)

model.fit(X, y_log)
y_pred_log = model.predict(X)

# convert back to original scale
y_pred_after = np.exp(y_pred_log)

print("R2 After (correct log use):", r2_score(y, y_pred_after))

plt.figure()
plt.scatter(X, y)
plt.plot(X, y_pred_after)
plt.title("After Log Transformation on y (Correct)")
plt.show()
```
