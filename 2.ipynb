```python
"""  Now taking data of simple Linear regression from a dataset from kaggle: salary vs experience dataset  """

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
z=pd.read_csv("C:\\Users\\mukes\\OneDrive\\Documents\\ML\\Experience-Salary.csv")
z
```

```python
x = z[["exp(in months)"]]
y = z[["salary(in thousands)"]]

# X = features, y = target
X_train, X_test, y_train, y_test = train_test_split(
    x, y,test_size=0.2,      # 20% test data
    random_state=42     # ensures same split every time
)

print(X_train.shape, X_test.shape) # get rows*columns size of both tran and test data
```

```python
model=LinearRegression()
model.fit(X_train,y_train)# Training model on the train dataset

y_pred=model.predict(X_test)#Predicting the target variable of the test dataset and then compare the original target variable of testing dataset and predicted dataset to know how model behaves on new unseen data

print("R2 Score:",r2_score(y_test,y_pred)) 
# R2 score here i can use as one of the availble tools to measure how much my y_pred vary from original y_test .
```
