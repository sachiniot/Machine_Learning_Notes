```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import PolynomialFeatures
```

```python
df=pd.read_csv('StudentPerformanceFactors.csv')
df.head(4)
```

```python
df.drop(columns=['Extracurricular_Activities','Internet_Access','Teacher_Quality','Parental_Education_Level','Distance_from_Home'],inplace=True)
df.head(4)
```

```python
cols = ['Parental_Involvement','Access_to_Resources','Motivation_Level','Family_Income','School_Type','Peer_Influence']


encoder1=OrdinalEncoder()
df[cols]=encoder1.fit_transform(df[cols])


encoder2 = OneHotEncoder(sparse_output=False)

encoded= encoder2.fit_transform(df[['Gender','Learning_Disabilities']])

encoded_df = pd.DataFrame(
    encoded,# has the encoded data stored
    columns=encoder2.get_feature_names_out(['Gender','Learning_Disabilities'])  # this will takes out columns names from cols
)
df = df.drop(columns=['Gender','Learning_Disabilities'])# will drop the original columns in df datset

df = pd.concat([df, encoded_df], axis=1)  # merging orginal datset with encoded data 

df.head(5)





```

```python
df.columns
```

```python
x=df.drop(columns=['Exam_Score'])
y=df['Exam_Score']
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)  # Enabling the interaction term
X_new = poly.fit_transform(x)
model1=LinearRegression()
model1.fit(X_new,y)
y_pred=model1.predict(X_new)
print("Mean Squared Error:",mean_squared_error(y,y_pred))
print("R2 Score:",r2_score(y,y_pred))
```

```python
# without Interaction term
model2=LinearRegression()
model2.fit(x,y)
y_pred=model2.predict(x)
print("Mean Squared Error:",mean_squared_error(y,y_pred))
print("R2 Score:",r2_score(y,y_pred))
```

```python
# Without interaction term,the model has a higher MSE and a Lower R2 score,indicating that the model with interaction term perform better in capturing the relationship between feature and target variable.
```
