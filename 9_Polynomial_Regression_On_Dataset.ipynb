```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import PolynomialFeatures
```

```python
df=pd.read_csv("Housing.csv")
df.head()
```

```python
df.columns
```

```python
cols = ['mainroad','guestroom','basement','hotwaterheating',
        'airconditioning','prefarea','furnishingstatus']

encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(df[cols])

encoded_df = pd.DataFrame(
    encoded,# has the encoded data stored
    columns=encoder.get_feature_names_out(cols)  # this will takes out columns names from cols
)
```

```python

df = df.drop(columns=cols)# will drop the original columns in df datset

df = pd.concat([df, encoded_df], axis=1)  # merging orginal datset with encoded data 
df.head()
```

```python
poly=PolynomialFeatures(degree=2,interaction_only=True,include_bias=False)
df_poly_converted_data=poly.fit_transform(df)
```
